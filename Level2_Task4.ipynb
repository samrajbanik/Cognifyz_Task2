```python
import pandas as pd

def analyze_restaurant_chains(df):
    """
    Identifies restaurant chains and analyzes their ratings and popularity.

    Args:
        df (pd.DataFrame): The input DataFrame containing restaurant data.

    Returns:
        None
    """

    # Count the occurrences of each restaurant name
    restaurant_counts = df['Restaurant Name'].value_counts()

    # Identify potential chains based on multiple occurrences
    potential_chains = restaurant_counts[restaurant_counts > 1]

    print("Potential restaurant chains:")
    print(potential_chains)

    # Analyze ratings and popularity of potential chains
    for chain in potential_chains.index:
        chain_data = df[df['Restaurant Name'] == chain]
        average_rating = chain_data['Aggregate rating'].mean()
        total_votes = chain_data['Votes'].sum()
        print(f"Chain: {chain}, Average Rating: {average_rating}, Total Votes: {total_votes}")

if __name__ == "__main__":
    # Replace 'your_data.csv' with the actual path to your dataset
    df = pd.read_csv("C:/Users/SAMRAJ/Downloads/Cognifyz Dataset.csv")
    analyze_restaurant_chains(df)
```
